```python
import tkinter 
from tkinter import *
import pickle
from tkinter import messagebox
import numpy as np
```

```python
root=Tk()

model=pickle.load(open("inventory.pkl","rb"))

def prediction():
    first_input=int(user_input1.get())
    sec_input=int(user_input2.get())
    third_input=int(user_input3.get())
    forth_input=int(user_input4.get())
    
    array=np.array([[first_input,sec_input,third_input,first_input]])
    value=model.predict(array)
    if value==0:
        messagebox.showinfo("Result","Your product Demand is very Bad")
    if value==1:
        messagebox.showinfo("Result","Your product Demand is very Fair")
    if value==2:
        messagebox.showinfo("Result","Your product Demand is very Good")
    if value==3:
        messagebox.showinfo("Result","Your product Demand is very Normal")
    
    
    
    
    
    
    






root.title(" GUI ")
root.geometry("500x500")
root.resizable(0,0)

tital=Label(root,text="INVENTORY PREDICTION GUI",fg="Black",background="white")
    tital.pack(ipady=10,pady=(5,0))
tital.config(font="Verdana")

tital1=Label(root,text="Enter Store Code",fg="Black",background="white")
tital1.pack(ipady=5,pady=(15,0))
tital1.config(font="Verdana")

user_input1=Entry(root)
user_input1.pack(ipady=5,pady=(5,0))




tital2=Label(root,text="Enter Product Code",fg="Black",background="white")
tital2.pack(ipady=5,pady=(15,0))
tital2.config(font="Verdana")

user_input2=Entry(root)
user_input2.pack(ipady=5,pady=(5,0))



tital3=Label(root,text="Enter Sales Quantity",fg="Black",background="white")
tital3.pack(ipady=5,pady=(15,0))
tital3.config(font="Verdana")

user_input3=Entry(root)
user_input3.pack(ipady=5,pady=(5,0))




tital4=Label(root,text="Enter Pending Order",fg="Black",background="white")
tital4.pack(ipady=5,pady=(15,0))
tital4.config(font="Verdana")

user_input4=Entry(root)
user_input4.pack(ipady=5,pady=(5,0))


login_but=Button(root,text="Get Prediction",bg="white",background="white",command=prediction)
login_but.pack(ipady=5,pady=(25,5))


root.mainloop()
```
